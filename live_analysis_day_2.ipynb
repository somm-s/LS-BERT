```python
import pandas as pd
column_names = ["timestamp", "ip1", "ip2", "prediction", "prob"]
path = 'live/predictions_777.csv'
df = pd.read_csv(path, names=column_names)
```

```python
df['prediction'] = 1 - df['prediction']
df['prob'] = 1 - df['prob']
```

```python
ip = "10.7.11.18"
df[df['ip2'] == ip]

# filter for entries younger than 1 hour, take current time
from datetime import datetime
now = datetime.now()
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df[df['timestamp'] > now - pd.Timedelta(minutes=15)]

df_sorted = df.sort_values(by='prob', ascending=False)
df_sorted.head(100)
```

```python
# for each ip1 and each ip2 separately sum predictions and output a sorted list by predictions for both ip1 and ip2
df1 = df.groupby('ip1')['prediction'].sum().reset_index()
df2 = df.groupby('ip2')['prediction'].sum().reset_index()
df1 = df1.sort_values(by='prediction', ascending=False)
df2 = df2.sort_values(by='prediction', ascending=False)
print(df1[df1['prediction'] > 0])
```
